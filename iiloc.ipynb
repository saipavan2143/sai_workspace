```

import pandas as pd
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Age":  [25, 30, 35],
    "City": ["Delhi", "Mumbai", "Chennai"]
})
print("DataFrame:\n", df, "\n")

```

```
print(df.iloc[0])
```

```
print(df.iloc[:, 1])
```

```
print(df.iloc[0:2])
```
